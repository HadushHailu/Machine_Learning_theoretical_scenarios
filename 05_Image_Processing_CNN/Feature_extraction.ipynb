## Feature Extraction
 - Corner Detection
 - Edge detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('stanford.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply corner detection (using the Harris corner detector)
corner_img = cv2.cornerHarris(img, 2, 3, 0.04)
corner_img = cv2.dilate(corner_img, None)
# plt.imshow(corner_img)
# plt.show()
```

```python
# Apply edge detection
edges = cv2.Canny(img, 100, 200)

# Apply blob detection
params = cv2.SimpleBlobDetector_Params()
params.filterByArea = True
params.minArea = 100
params.filterByCircularity = False
params.filterByConvexity = False
params.filterByInertia = False
```

```python
detector = cv2.SimpleBlobDetector_create(params)
keypoints = detector.detect(img)
```

```python
# Display the results
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(corner_img)
plt.title('Corners')

plt.subplot(1, 3, 3)
plt.imshow(edges)
plt.title('Edges')

plt.show()
```
